### 계열 매출 데이터를 활용한 월별 매출 분석

```python
# 필요한 모듈 불러오기

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
# 한글 폰트 설정 (윈도우 기준)
font_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
class SalesAnalysis:
    def __init__(self):
        # 날짜 생성 (2024년 1월~12월)
        self.dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
        # 난수 시드 고정
        np.random.seed(42)
        # 난수 매출 생성 (1000~10000)
        self.sales = np.random.randint(1000, 10001, size=len(self.dates))
        # 데이터프레임 생성
        self.data = pd.DataFrame({'Date': self.dates, 'Sales': self.sales})

    def monthly_sales(self):
        # 월 정보 추가
        self.data['Month'] = self.data['Date'].dt.to_period('M')
        # 월별 매출 합계 계산
        monthly_sum = self.data.groupby('Month')['Sales'].sum()
        return monthly_sum

    def plot_monthly_sales(self):
        # 월별 매출 데이터 가져오기
        monthly_sum = self.monthly_sales()
        # 그래프 크기 설정
        plt.figure(figsize=(10,6))
        # 꺾은선 그래프 그리기
        plt.plot(monthly_sum.index.astype(str), monthly_sum.values, marker='o', linestyle='-')
        # 제목 설정
        plt.title("2024년 월별 매출 총합")
        # x축 레이블
        plt.xlabel("월")
        # y축 레이블
        plt.ylabel("매출 합계 (원)")
        # 격자 표시
        plt.grid(True)
        # x축 눈금 회전
        plt.xticks(rotation=45)
        # 그래프 출력
        plt.show()
```

```python
analyzer = SalesAnalysis()          # 클래스 객체 생성
print(analyzer.monthly_sales())     # 월별 매출 합계 출력
analyzer.plot_monthly_sales()       # 월별 매출 그래프 출력
```
